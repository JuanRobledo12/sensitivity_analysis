```python
df_feature_imp = pd.read_csv('feature_importances_per_target_var.csv')
df_feature_imp.head(20)
```

```python
df_feature_imp.tail(20)
```

```python
# Add a rank index for each Subsector
df_feature_imp['Rank'] = df_feature_imp.groupby('Subsector').cumcount() + 1

# Pivot the table
pivoted_df = df_feature_imp.pivot(index='Rank', columns='Subsector', values='Feature')

# Reset index for cleaner output
pivoted_df.reset_index(inplace=True)

# Save to CSV
pivoted_df.to_csv('pivoted_features.csv', index=False)

print(pivoted_df)
```
